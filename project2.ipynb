```python
#Bar Plot
```

```python
data["TotalPrice"].hist(bins=25, color="skyblue",
    edgecolor="black")
plt.title("Total Price Distribution")
plt.xlabel("Total Price")
plt.ylabel("Number of Orders")
plt.show()
```

```python
data["PaymentMethod"].value_counts().plot(
    kind="bar",
    color="skyblue",
    edgecolor="black"
)

plt.title("Payment Method Distribution")
plt.xlabel("Payment Mode")
plt.ylabel("Number of Orders")
plt.show()
```

```python
import matplotlib.pyplot as plt

data["OrderStatus"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Order Status Distribution")
plt.ylabel("")  # Y-axis label remove karne ke liye
plt.show()
```

```python
import matplotlib.pyplot as plt

data.groupby(data["Date"].dt.month)["TotalPrice"].sum().plot(
    kind="line",
    marker="o"
)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True)

plt.show()
```
